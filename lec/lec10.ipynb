```python
galton = Table.read_table('galton.csv')
galton
```

```python
galton.scatter('midparentHeight', 'childHeight')
```

```python
galton.scatter('midparentHeight', 'childHeight')
plots.plot([67.5, 67.5], [50, 85], color='red', lw=2)
plots.plot([68.5, 68.5], [50, 85], color='red', lw=2);
```

```python
nearby = galton.where('midparentHeight', are.between(67.5, 68.5))
nearby.column('childHeight').mean()
```

```python
galton.scatter('midparentHeight', 'childHeight')
plots.plot([67.5, 67.5], [50, 85], color='red', lw=2)
plots.plot([68.5, 68.5], [50, 85], color='red', lw=2)
plots.scatter(68, 66.24, color='gold', s=50);
```

```python
def predict(h):
    nearby = galton.where('midparentHeight', are.between(h- 1/2, h+ 1/2))
    return nearby.column('childHeight').mean()
```

```python
predict(68)
```

```python
predict(70)
```

```python
predict(73)
```

```python
predicted_heights = galton.apply(predict, 'midparentHeight')
```

```python
galton = galton.with_column('predictedHeight', predicted_heights)
```

```python
galton.select('midparentHeight', 'predictedHeight', 'childHeight').scatter('midparentHeight')
```

```python
def difference(x,y):
    return x-y
```

```python
prediction_errors = galton.apply(difference, 'childHeight', 'predictedHeight')
```

```python
galton = galton.with_column('errors', prediction_errors)
galton
```

```python
galton.hist('errors', group = 'gender') #creates two different groups on the same graph
```
